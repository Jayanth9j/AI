```python
import pandas as pd 

hospital= 'C:\\Users\\Jayanth\\OneDrive - The Pennsylvania State University\\Desktop\\Visual\\python\\processed_hospital_dataset.csv'

df = pd.read_csv(hospital)
```

```python
df = df.sort_values(by='age')
df.reset_index(drop=True, inplace=True)
```

```python
print(df.head())
```

```python
high_systolic = df[df['blood_pressure'].str.contains(r'^(?:1[4-9]\d|[2-9]\d{2})/', na=False)] 
```

```python
df['gender'].str.contains(r'^male$', case=False, na=False) # Checks if the entire string is "male", ignoring case. Returns a boolean Series.
```

```python
df.info()
```

```python
# patient_id to string
df['patient_id'] = df['patient_id'].astype(str)
```

```python
# Fill missing values in numerical columns (continuous - interval/ratio)
df['age'] = df['age'].fillna(df['age'].median())             # age: numerical (interval)
df['cholesterol'] = df['cholesterol'].fillna(df['cholesterol'].median())  # cholesterol: numerical (ratio)
df['bmi'] = df['bmi'].fillna(df['bmi'].median())             # bmi: numerical (ratio)

# Fill missing values in categorical columns (nominal)
for col in ['gender', 'diabetes', 'hypertension', 'discharge_destination', 'readmitted_30_days']:
    df[col] = df[col].fillna(df[col].mode()[0])              # All: categorical (nominal)
```

```python
# Drop rows missing patient_id (identifier - nominal)
df.dropna(subset=['patient_id'], inplace=True)
```

```python
df
```

```python
# Encode binary nominal variables from Yes/No to 1/0
binary_map = {'Yes': 1, 'No': 0}
df['diabetes'] = df['diabetes'].map(binary_map)             # diabetes: binary categorical (nominal)
df['hypertension'] = df['hypertension'].map(binary_map)     # hypertension: binary categorical (nominal)
df['readmitted_30_days'] = df['readmitted_30_days'].map(binary_map)  # readmitted: binary categorical (nominal)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1 Histogram - Distribution of Age
# Variable type: age = numerical (interval)
plt.figure(figsize=(6, 4))
sns.histplot(df['age'], kde=True, bins=30)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 2 Boxplot - BMI across Gender
# Variable types: bmi = numerical (ratio), gender = categorical (nominal)
plt.figure(figsize=(6, 4))
sns.boxplot(x='gender', y='bmi', data=df)
plt.title("BMI Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("BMI")
plt.show()
```

```python
# 3 Countplot - Frequency of Discharge Destinations
# Variable type: discharge_destination = categorical (nominal)
plt.figure(figsize=(6, 4))
sns.countplot(x='discharge_destination', data=df)
plt.title("Discharge Destination Counts")
plt.xlabel("Discharge Destination")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5 Scatterplot - BMI vs Cholesterol colored by Diabetes
# Variable types: bmi, cholesterol = numerical (ratio), diabetes = binary categorical (nominal)
plt.figure(figsize=(6, 4))
sns.scatterplot(x='bmi', y='cholesterol', hue='diabetes', data=df)
plt.title("BMI vs Cholesterol by Diabetes Status")
plt.xlabel("BMI")
plt.ylabel("Cholesterol")
plt.legend(title='Diabetes')
plt.show()
```
