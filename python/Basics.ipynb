```python
from functools import reduce
# 1. Arithmetic: double length_of_stay
df['LOS_Double'] = df['length_of_stay'] * 2  # Arithmetic: *

# 2. Comparison: stays > 5 days?
df['LOS_GT_5'] = df['length_of_stay'] > 5  # Comparison: >

# 3. Logical: seniors with long stays
df['Senior_LongStay'] = (df['age'] > 60) & (df['length_of_stay'] > 7)  # Logical: &

# 4. Assignment: total medications
total_meds = 0            # Assignment: =
total_meds += df['medication_count'].sum()  # Assignment: +=

# 5. Bitwise: demo on age
df['BitwiseDemo'] = df['age'].astype(int) & 1  # Bitwise: &

# 6. Membership: discharge to home or rehab?
df['IsHomeDischarge'] = df['discharge_destination'].isin(['Home','Rehab'])  # Membership: in

# 7. Identity: compare objects
a = df
b = df.copy()
same_object = (a is df)            # Identity: is
different_object = (b is not df)   # Identity: is not

# 8. Ternary: label length of stay
df['LOS_Label'] = df['length_of_stay'].apply(
    lambda x: 'Short' if x <= 5 else 'Long'   # Ternary: if-else
)

# 9. Sorting
sorted_by_age       = df.sort_values(by='age')                             # ascending
sorted_by_meds_desc = df.sort_values(by='medication_count', ascending=False)  # descending

# 10. map/filter/reduce
df['AgeSquared']   = df['age'].map(lambda x: x**2)                         # map()
filtered_df        = df[df['medication_count'] > 5]                        # filter
total_bmi          = reduce(lambda x, y: x + y, df['bmi'])                 # reduce()

print("Total meds:", total_meds)
print("Filtered DF (meds > 5):", filtered_df.shape)
```

```python
import pandas as pd

# ── 1. LIST: ages ─────────────────────────────────────────────────────────
age_list = df['age'].tolist()                      # Identifier: age_list

# Indexing (first element)
first_age = age_list[0]                            # first_age = age of the first patient

# Slicing (elements 2–5)
slice_ages = age_list[1:5]                         # slice_ages = ages of patients 2–5

# Reindexing (custom order)
new_order = [4, 0, 3, 1, 2]                        # pick indices 5th,1st,4th,2nd,3rd
reindexed_list = [age_list[i] for i in new_order]  # reindexed_list holds ages in new order


# ── 2. STRING: discharge_destination of first patient ────────────────────
dest = df.loc[0, 'discharge_destination']          # Identifier: dest (e.g. "Home")

# Indexing (first character)
first_char = dest[0]                               # first_char = first letter of dest

# Slicing (chars 2–4)
slice_str = dest[1:4]                              # slice_str = substring from index 1 to 3

# Reindexing (reverse)
reindexed_str = "".join(dest[i] for i in range(len(dest)-1, -1, -1))  # reversed string


# ── 3. TUPLE: medication counts ──────────────────────────────────────────
meds_list = df['medication_count'].tolist()        # convert to list first
meds_tuple = tuple(meds_list)                      # Identifier: meds_tuple

# Indexing
second_meds = meds_tuple[1]                        # second_meds = 2nd patient's med count

# Slicing
slice_meds = meds_tuple[2:5]                       # slice_meds = med counts 3–5

# Reindexing (rotate left by one)
rotated_tuple = meds_tuple[1:] + meds_tuple[:1]    # rotated meds tuple


# ── 4. DICT: patient_id → readmitted_30_days for first 5 ───────────────
keys   = df['patient_id'][:5].tolist()             # first 5 IDs
values = df['readmitted_30_days'][:5].tolist()     # corresponding readmission flags
dict_example = dict(zip(keys, values))             # Identifier: dict_example

# Indexing by key
val_for_first = dict_example[keys[0]]              # lookup readmission for first patient

# Slicing keys (first 3)
slice_keys = list(dict_example.keys())[:3]         # slice of first 3 patient IDs

# Reindexing dict (new key order)
new_key_order = slice_keys[::-1]                   # reverse the first 3 IDs
reindexed_dict = {k: dict_example[k] for k in new_key_order}  # dict in new order


# ── Print Results ────────────────────────────────────────────────────────
print("List indexing:", first_age)
print("List slicing:", slice_ages)
print("List reindexed:", reindexed_list)

print("String indexing:", first_char)
print("String slicing:", slice_str)
print("String reindexed:", reindexed_str)

print("Tuple indexing:", second_meds)
print("Tuple slicing:", slice_meds)
print("Tuple reindexed:", rotated_tuple)

print("Dict indexing:", val_for_first)
print("Dict slicing keys:", slice_keys)
print("Dict reindexed:", reindexed_dict)
```
